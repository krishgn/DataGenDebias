```python
import pandas as pd
import numpy as np
from tgan.model import TGANModel
```

```python
data = pd.read_csv('../data/processed_female.csv',low_memory=False)
print(data.info())
continuous_columns = [0, 2, 3 ,9, 10, 11]
```

```python
tgan = TGANModel(continuous_columns)
tgan.fit(data)
```

```python
num_samples = 1000
samples = tgan.sample(num_samples)
samples.head(3)
```

```python
samples.to_csv('../data/generated_female_1000.csv')
```

```python
model_path = '../demos/Model2.pkl'
tgan.save(model_path)
```

```python
new_tgan = TGANModel.load(model_path)
new_samples = new_tgan.sample(1000)
new_samples.head(3)
```
